```python
import numpy as np
from scipy.interpolate import BSpline
import matplotlib.pyplot as plt
```

Construct a quadratic (degree p=2) spline function on the base interval 
of 2 <= x <= 4.

```python
p = 2 # degree
t = [0, 0, 0, 3, 4, 5, 6] # knot vector
# c = [-1, 2, 0, -1] # spline coefficients
c = [1, 0, 0, 0] # spline coefficients
bsp = BSpline(t, c, p)
bsp(2.5)
fig, ax = plt.subplots()
# xx = np.linspace(1.5, 4.5, 10)
xx = np.linspace(0, 2, 10)
ax.plot(xx, bsp(xx), 'bo-', lw=2, label='BSpline')
ax.grid(True)
ax.legend(loc='best')
plt.show()
```

```python
c0 = [0, 1] # size >= n, n >= p+1
# len(c0) = number_of_control_points = (n+1) => n = len(c0) - 1
p0 = 0 # degree 0, constant
# len(kv) = (p+1) + len(c0)
kv0 = [0, 1, 2]
B00 = BSpline(kv0, c0, p0)
fig, ax = plt.subplots()
ax.plot(xx, B00(xx), 'bo-', lw=2, label='BSpline')
ax.grid(True)
ax.legend(loc='best')
plt.show()
```

```python
nti = 2 # number of time intervals

ncp = 5 # number of control points

p = 1 # degree 1, linear

kv = [0, 1, 2, 3, 4, 5, 6] # knot vector, len(kv) = (p+1) + len(c)
# kv = [0, 0, 1, 2, 3, 4, 4] # knot vector, len(kv) = (p+1) + len(c)

npoints = (kv[-1]) * 2**nti

xx = np.linspace(0, kv[-1], npoints, endpoint=False)

for j in np.arange(ncp):
    
    c = np.zeros(ncp)
    c[j] = 1.0 
    
    B = BSpline(kv, c, p, extrapolate=False)
    
    fig, ax = plt.subplots()
    
    ax.plot(xx, B(xx), 'bo', lw=2, label=f'$N_{j}^{p}$')
    ax.grid(True)
    ax.legend(loc='best')
    plt.show()
```

```python
nti = 4 # number of time intervals
ncp = 2 # number of control points

p = 1 # degree

kv = [0, 0, 1, 1] # knot vector, len(kv) = (p+1) + len(c)

npoints = (kv[-1]) * 2**nti + 1

xx = np.linspace(0, kv[-1], npoints, endpoint=True)
N = []

for j in np.arange(ncp):
    
    c = np.zeros(ncp)
    c[j] = 1.0 
    
    B = (BSpline(kv, c, p, extrapolate=False))
    N.append(np.nan_to_num(B(xx), copy=False, nan=0.0))
    
    
fig, ax = plt.subplots()

for j in np.arange(ncp):
    
    ax.plot(xx, N[j], 'o-', lw=2, label=f'$N_{j}^{p}$')

ax.grid(True)
ax.axis('equal')
# ax.legend(loc='best')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```
